```python
import pandas as pd
import numpy as np
import requests
from tqdm import tqdm
import uuid
import PyPDF2
import pdfplumber
from io import BytesIO
data = pd.read_csv("data/cleanOutput.csv", encoding='utf-8-sig')
data.shape
```

```python
data["fileName"] = ""
kübPdfFilePath="data/pdf/küb/"
ktPdfFilePath="data/pdf/kt/"

kübTxtFilePath="data/txt/küb/"
ktTxtFilePath="data/txt/kt/"

errorsHttpCode = []
textZero = [] 
```

```python
#pdf dowload
for index, row in tqdm(data.iterrows(), total=len(data), desc="Processing rows"):
    break
    kübUrl=row["kübDocLink"]
    ktUrl=row["ktDocLink"]
    fileName = str(uuid.uuid1())+".pdf"

    data.at[index, "fileName"] = fileName

    if not pd.isna(kübUrl):

        response = requests.get(kübUrl)
        with open(kübPdfFilePath+fileName, "wb") as f:
            f.write(response.content)


    if not pd.isna(ktUrl):

        response = requests.get(ktUrl)
        with open(ktPdfFilePath+fileName, "wb") as f:
            f.write(response.content)
    
    
    
```

```python
#pd2

for  index, row in tqdm(data.iterrows(), total=len(data), desc="Processing rows"):
    break
    kübUrl=row["kübDocLink"]
    ktUrl=row["ktDocLink"]

    kübUrl = "https://titck.gov.tr/storage/kubKtAttachments/eb17cd6b60137.pdf"
    ktUrl = "https://titck.gov.tr/storage/kubKtAttachments/f9960b6943243.pdf"

    fileName = str(uuid.uuid1())+".txt"
    data.at[index, "fileName"] = fileName

    if not pd.isna(kübUrl):

        response = requests.get(kübUrl)
        if response.status_code == 200:
            
            pdf_file = BytesIO(response.content)
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            text = ''
            
            
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                text += page.extract_text()
            
            
            if len(text)>0:
                with open(kübTxtFilePath+fileName, 'w', encoding='utf-8') as text_file:
                    text_file.write(text)
            else:
                print("zero text ")
                textZero.append({
                    "index":index,
                    "type":"küb"
                })
            

        else:
            print(f"PDF dosyasına ulaşılamadı. HTTP Durum Kodu: {response.status_code}")
            errorsHttpCode.append({
                "index":index,
                "type":"küb"
            })




    if not pd.isna(ktUrl):

        response = requests.get(ktUrl)
        if response.status_code == 200:
           
            pdf_file = BytesIO(response.content)
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            text = ''
            
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                text += page.extract_text()
            
            if len(text)>0:
                with open(ktTxtFilePath+fileName, 'w', encoding='utf-8') as text_file:
                    text_file.write(text)
            else:
                print("zero text ")
                textZero.append({
                    "index":index,
                    "type":"kt"
                })
            

        else:
            print(f"PDF dosyasına ulaşılamadı. HTTP Durum Kodu: {response.status_code}")
            errorsHttpCode.append({
                "index":index,
                "type":"kt"
            })

    break
    
    
    
```

```python
#pdfplumber 
#image olarak yükjlenmisse zero text e ekleniyor 

for index, row in tqdm(data.iterrows(), total=len(data), desc="Processing rows"):
    break
    kübUrl = row["kübDocLink"]
    ktUrl = row["ktDocLink"]

    kübUrl="https://titck.gov.tr/storage/Archive/2024/kubKtAttachments/UYGUNKBLEVOKAST_8b677797-094c-4da5-89d4-adca150e9030.pdf"
    ktUrl="https://titck.gov.tr/storage/Archive/2020/kubKtAttachments/KTLEVOKAST_b9e7b6ef-5383-4f69-b04c-5d711d6ff169.pdf"
    
    fileName = str(uuid.uuid1()) + ".txt"
    data.at[index, "fileName"] = fileName

    
    if not pd.isna(kübUrl):
        response = requests.get(kübUrl)
        if response.status_code == 200:
            pdf_file = BytesIO(response.content)

            
            with pdfplumber.open(pdf_file) as pdf:
                text = ""
                for page in pdf.pages:
                    text += page.extract_text() or ""  # Sayfadan metni çıkar

            if len(text)>0:
                with open(kübTxtFilePath + fileName, 'w', encoding='utf-8') as text_file:
                    text_file.write(text)
            else:
                print("zero text ")
                textZero.append({
                    "index":index,
                    "type":"küb"
                })
                
        else:
            print(f"Küb PDF dosyasına ulaşılamadı. HTTP Durum Kodu: {response.status_code}")
            errorsHttpCode.append({
                "index":index,
                "type":"küb"
            })

    
    

    



    if not pd.isna(ktUrl):
        response = requests.get(ktUrl)
        
        if response.status_code == 200:
            pdf_file = BytesIO(response.content)

            
            with pdfplumber.open(pdf_file) as pdf:
                text = ""
                for page in pdf.pages:
                    text += page.extract_text() or ""  # Sayfadan metni çıkar

            if len(text)>0:
                with open(ktTxtFilePath + fileName, 'w', encoding='utf-8') as text_file:
                    text_file.write(text)
            else:
                print("zero text ")
                textZero.append({
                    "index":index,
                    "type":"kt"
                })

        else:
            print(f"KT PDF dosyasına ulaşılamadı. HTTP Durum Kodu: {response.status_code}")
            errorsHttpCode.append({
                "index":index,
                "type":"kt"
            })

    break
```

```python
#pdfplumber for küb
#image olarak yükjlenmisse zero text e ekleniyor
kübErrors = [] 
for index, row in tqdm(data.iterrows(), total=len(data), desc="Processing kübs"):
    
    kübUrl = row["kübDocLink"]
    
    fileName = str(uuid.uuid1()) + ".txt"
    data.at[index, "fileName"] = fileName

    
    if not pd.isna(kübUrl):
        response = requests.get(kübUrl)
        if response.status_code == 200:
            pdf_file = BytesIO(response.content)
            text = ""
            try:
                with pdfplumber.open(pdf_file) as pdf:
                    
                    for page in pdf.pages:
                        text += page.extract_text() or "" 


                if len(text)>0:
                    with open(kübTxtFilePath + fileName, 'w', encoding='utf-8') as text_file:
                        text_file.write(text)
                else:
                    print("zero text ")
                    kübErrors.append({
                        "index":index,
                        "reason":"zero text"
                    })

            except Exception as e: 
                print("Error:", str(e))  
                kübErrors.append({
                    "index": index,
                    "reason": "pdf read",
                    "error_message": str(e)
                })
                

        else:
            print(f"Küb PDF dosyasına ulaşılamadı. HTTP Durum Kodu: {response.status_code}")
            kübErrors.append({
                "index":index,
                "reason":"http status"
            })

    
```
